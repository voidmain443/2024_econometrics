우리는 이런 합집합 교집합 차집함 등을 하나의 Set이라는 함수를 만들어볼 수 있을 것입니다. 
이미 파이썬에는 구현되어 있는 기능이지만 내부적으로 직접 클래스를 만들어 함수를 직접 필요한 연산을 수행하는 하나의 큰 클래스를 만드는 것은 의미 있습니다. 

있는 Set을 쓰는 것도 현명한 방법이지만. 미시경제학에서 우리가 앞으로 정의할 인간의 욕망의 정의를 어떤 형태로 하느냐에 따라서 우리는 교환이라는 개념자체를 혹은 무차별 곡선의 전제조건 자체를 코딩해야할 수도 있습니다. 
경제학원론에서 배웠던 모든 함수들에는 1명 및 금액의 단위 등을 고려하고 받아들이는 학생이 많지는 않을 것으로 생각합니다. 

따라서 Set이라는 큰 개념 속에서 교집합, 차집합등이 존재하듯 . 

인간의 행동원리라고 우리가 믿고 있는 욕망의 단위 및 표현 들도 이처럼 존재할 수 있습니다. 

Set{교집합, 차집합 ...etc}
인간의 욕망 {선호, 현시선호, 절대선호, .. etc}
의 형태로 먼저 스스로 클래스를 만들어 쓰는 공부가 잘 되어있어야합니다.

````python
class Set_class():
    def setdata(self,a,b):
        self.a = a 
        self.b = b
        
    def union(self):
        result = self.b + []
        for i in self.a:
            if i not in self.b:
                result.append(i)
        result.sort()
        return result
    
    def intersection(self):
        result = []
        for i in self.a:
            if i in self.b:
                result.append(i)
        result.sort()
        return result
    
    def complement(self):
        result = self.a + []
        for i in self.b:
            if i in self.a:
                result.remove(i)
        result.sort()
        return result
````

````python
c = Set_class()
c.setdata(a,b)
print(c.a,c.b)

print ('{}와 {}의 합집합은 {}이다.'.format(c.a, c.b, c.union()))         # c라는 객체의 변수 a,b를 가지고 합집합을 구한다.
print ('{}와 {}의 교집합은 {}이다.'.format(c.a, c.b, c.intersection()))  # c라는 객체의 변수 a,b를 가지고 교집합을 구한다.
print ('{}와 {}의 차집합은 {}이다.'.format(c.a, c.b, c.complement()))
````

derivatives


````python
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m


def my_function(x):
    return x**2

slope_at_2 = derivative_x(my_function, 2, .00001)

print(slope_at_2) # prints 4.000010000000827
````

Partial Derivatives

$$ f(x,y) = 2x^3 + 3y^3$$
$$ \frac{d}{dx}2x^3 + 3y^3 = 6x^3 $$
$$ \frac{d}{dy}2x^3 + 3y^3 = 9y^2 $$

````python
from sympy import *
from sympy.plotting import plot3d

# Declare x and y to SymPy
x,y = symbols('x y')

# Now just use Python syntax to declare function
f = 2*x**3 + 3*y**3

# Calculate the partial derivatives for x and y
dx_f = diff(f, x)
dy_f = diff(f, y)

print(dx_f) # prints 6*x**2
print(dy_f) # prints 9*y**2

# plot the function
plot3d(f)
````
